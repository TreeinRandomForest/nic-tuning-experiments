### Question 1: Is time dependent on the number of instructions?

```python
plt.figure(figsize=(7,7))
plt.plot(df_comb.ins_total_mean, df_comb.time_mean, 'p')

df2 = df_comb.copy()
df2.dvfs = df2.dvfs.apply(lambda x: int(x, base=16))

df3 = df2[df2.sys=='linux']
plt.plot(df3.ins_total_mean, df3.time_mean, 'p', c='y', label='linux')
df3 = df2[df2.sys=='ebbrt']
plt.plot(df3.ins_total_mean, df3.time_mean, 'p', c='g', label='ebbrt')
plt.legend()

plt.xlabel('Instructions Total')
plt.ylabel('Time Taken')
```
